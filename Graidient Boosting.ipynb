# Gradient Boost Classifier

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('brca.csv')
df.head()
```

```python
df.info()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['y']=label.fit_transform(df['y'])
```

```python
df.head()
```

```python
X=df.drop('y',axis=1)
y=df['y']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.model_selection import KFold
fold=KFold(n_splits=100,random_state=42,shuffle=True)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
cv=cross_val_score(model,X_train,y_train,cv=fold)

for i in range(len(cv)):
    print("model",i,'accuracy:',cv[i])
```

```python
from sklearn.metrics import classification_report
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
```

# Gradient Regressor

```python
df=pd.read_csv('Life Expectancy.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Country']=label.fit_transform(df['Country'])
df['Year']=label.fit_transform(df['Year'])
df['Status']=label.fit_transform(df['Status'])
```

```python
df.head()
```

```python
df['Country'].unique()
```

```python
df['Alcohol'].unique()
```

```python
df.isnull().sum()
```

```python
df['Hepatitis B'].unique()
```

```python
df['Schooling'].unique()
```

```python
df['BMI']=df[' BMI ']
```

```python
a=['Life expectancy ','Adult Mortality','Alcohol','Hepatitis B','GDP','Polio','Schooling','Diphtheria ']
b=[' BMI ','Total expenditure','Population',' thinness  1-19 years',' thinness 5-9 years','Income composition of resources']
for i in a:
    df[i].fillna(df[i].mode()[0],inplace=True)
for i in b:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
X=df.drop('Life expectancy ',axis=1)
y=df['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.model_selection import KFold
fold=KFold(random_state=7,n_splits=100,shuffle=True)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor(n_estimators=10,learning_rate=0.1,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
cv=cross_val_score(model,X_train,y_train,cv=fold)
for i in range(len(cv)):
    print("model:",i,'accuracy:',cv[i])
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(mean_absolute_error(y_test,y_pred))
```
