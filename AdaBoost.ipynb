# AdaBoost Classifier

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.info()
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Sex']=label.fit_transform(df['Sex'])
df['Cabin']=label.fit_transform(df['Cabin'])
df['Embarked']=label.fit_transform(df['Embarked'])
```

```python
df.info()
```

```python
X=df.drop(['Survived','Name','Ticket'],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
base_model=SVC(kernel='rbf')
bmodel=DecisionTreeClassifier()
```

```python
from sklearn.naive_bayes import GaussianNB
cmodel=GaussianNB()
```

```python
from sklearn.model_selection import KFold
fold=KFold(n_splits=100,random_state=7,shuffle=True)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier(base_estimator=cmodel,n_estimators=10,random_state=7)
```

```python
from sklearn.model_selection import cross_val_score
cv=cross_val_score(model,X_train,y_train,cv=fold)
for i in range(len(cv)):
    print('model:',i,'accuracy:',cv[i])
```

```python
from sklearn.metrics import classification_report
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
```

## Adaboost Regressor


```python
df=pd.read_csv('auto-mpg.csv')
df.head()
```

```python
df.info()
```

```python
X=df.drop(['car name','mpg'],axis=1)
y=df['mpg']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeRegressor
amodel=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import KFold
Fold=KFold(n_splits=100,random_state=7,shuffle=True)
```

```python
from sklearn.ensemble import AdaBoostRegressor
model=AdaBoostRegressor(estimator=amodel,n_estimators=10,random_state=7)
```

```python
from sklearn.model_selection import cross_val_score
cv=cross_val_score(model,X_train,y_train,cv=Fold)
for i in range(len(cv)):
    print('model:',i,'accuracy:',cv[i])
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
