## Digits Recognition

```python
import pandas as pd 
from sklearn.datasets import load_digits
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits=load_digits
```

```python
dir(digits)
```

```python
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i])
plt.show()
```

```python
digits.feature_names
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('target',axis=1)
y=df['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=70)
model.fit(X_train,y_train)
```

```python
print(RandomForestClassifier.get_params)
```

```python
model.score(X_test,y_test)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,pred)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(confusion_matrix(y_test,pred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

### Cancer Dataset

```python
df=pd.read_csv('Cancer_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
df['diagnosis']=label.fit_transform(df['diagnosis'])
```

```python
df['diagnosis'].unique()
```

```python
df.info()
```

```python
X=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
model.score(X_test,y_test)
```

```python
for i in range(len(model.feature_importances_)):
    print(df.columns[i],":",model.feature_importances_[i])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def tunning(X,y):
    from sklearn.model_selection import GridSearchCV
    from sklearn.ensemble import RandomForestClassifier
    model=RandomForestClassifier
    param_grid={'criterion':['gini', 'entropy', 'log_loss'],'max_depth':[5,10,15,20,25],'max_features':['sqrt','log'],'n_estimators':[50,100,150,200]}
    grid=GridSearchCV(param_grid=param_grid,cv=5,estimator=RandomForestClassifier())
    grid.fit(X,y)
    print("Best Parameters",grid.best_params_)
    print("Best Score",grid.best_score_)
    
```

```python
tunning(X_train,y_train)
```

```python
model1=RandomForestClassifier(criterion='log_loss',max_depth=5,max_features='sqrt',n_estimators=200)
model1.fit(X_train,y_train)
```

```python
pred1=model1.predict(X_test)
```

```python
confusion_matrix(y_test,pred1)
```

```python
print(classification_report(y_test,pred1))
```
